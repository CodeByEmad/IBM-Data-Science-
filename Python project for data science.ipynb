```python
# Import necessary libraries
import yfinance as yf
import pandas as pd

# Extracting Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")  # Fetch historical data
tesla_data.reset_index(inplace=True)  # Reset the index

# Display the first five rows
print(tesla_data.head())
```

```python
import matplotlib.pyplot as plt

# Plotting Tesla stock closing prices
plt.figure(figsize=(14, 7))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla (TSLA)', color='blue')
# Formatting the plot
plt.title('Stock Closing Prices: TSLA vs GME')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Extracting GameStop stock data
gamestop = yf.Ticker("GME")
gme_data = gamestop.history(period="max")  # Fetch historical data
gme_data.reset_index(inplace=True)  # Reset the index

# Display the first five rows
print(gme_data.head())
```

```python
import matplotlib.pyplot as plt

# Function to plot Tesla stock data
def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Date'], data['Open'], label='Open Price', color='blue')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

# Create a graph for Tesla stock data
make_graph(tesla_data, "Tesla Stock Prices")
```

```python
# Function to plot GameStop stock data
def make_graph_gme(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Date'], data['Open'], label='Open Price', color='green')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

# Create a graph for GameStop stock data
make_graph_gme(gme_data, "GameStop Stock Prices")
```
